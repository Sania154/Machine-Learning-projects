# relplot()

```python
df.head()
```

```python
sns.relplot(x='age',y='fare',col='pclass',data=df,hue='sex',style='sex',kind='line')
plt.show()
```

# scatter plot

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x='age',y='fare',hue='sex',data=df)
plt.show()
```

# 3 categorical plot

# 3.1 Barplot

```python
plt.figure(figsize=(8,8))
sns.barplot(x='sex',y='survived',hue='pclass',data=df)
plt.show()
```

# 3.2 stripplot

```python
plt.figure(figsize=(8,8))
sns.stripplot(x='sex',y='age',hue='sex',data=df)
plt.show()
```

# 3.3 Swarmplot

```python
plt.figure(figsize=(8,8))
sns.swarmplot(x='sex',y='age',hue='sex',data=df)
plt.show()
```

# 3.4 Boxplot

```python
plt.figure(figsize=(8,8))
sns.boxplot(y='age',x='survived',data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(y='age',x='survived',hue='sex',data=df)
plt.show()
```

# 3.5 Violinplot

```python
plt.figure(figsize=(8,8))
sns.violinplot(y='age',x='survived',hue='sex',data=df)
plt.show()
```

# 3.6 Countplot

```python
plt.figure(figsize=(8,8))
sns.countplot(x='survived',data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.countplot(x='survived',hue='sex',data=df)
plt.show()
```

# 3.7 pointplot

```python
plt.figure(figsize=(8,8))
sns.pointplot(y='survived',x='sex',hue='pclass',data=df)
plt.show()
```

# 4 Regression plots

# 4.1 Implot()

```python
sns.lmplot(x='age', y='fare', data=df)
plt.show()
# not best correlation
```

# 4.2 Regplot

```python
sns.regplot(x='fare', y='fare', data=df)
plt.show()
#strong correlation
```

# 5 Matrix plots

# 5.1 Heatmap

```python
plt.subplots(figsize=(10,10))
sns.heatmap(df.corr(),cmap='coolwarm',annot=True,fmt='2f')
plt.show()
```
